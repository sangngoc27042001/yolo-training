```python
from ultralytics import YOLO

# Load the YOLO11 model
model = YOLO("yolo11n-pose.pt")

# Export the model to TFLite format with optimizations for browser use
# Key optimizations:
# 1. imgsz=320: Reduces input from 640x640 to 320x320 (4x fewer detections)
# 2. int8=True: INT8 quantization for smallest model size
# 3. nms=True: Include NMS in model to filter detections (reduces output dramatically)

model.export(
    format="tflite",
    # imgsz=320,     # Smaller input = fewer detections (from 8400 to ~2100)
    int8=True,     # INT8 quantization for smallest size
    nms=True,      # Include Non-Maximum Suppression to filter detections
    simplify=True,  # Simplify the model graph
)

# Alternative: Float16 with NMS (if INT8 accuracy is insufficient)
# model.export(
#     format="tflite",
#     imgsz=320,
#     half=True,
#     nms=True,
#     simplify=True
# )

# Load the exported TFLite model
tflite_model = YOLO("yolo11n-pose_saved_model/yolo11n-pose_int8.tflite")

# Run inference with smaller input size
results = tflite_model("https://ultralytics.com/images/bus.jpg", imgsz=320)
```
