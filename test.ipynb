```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLO26n model
model = YOLO("yolo26n.pt")

model.export(
    format="tflite",
    # imgsz=320,     # Smaller input = fewer detections (from 8400 to ~2100)
    int8=True,     # INT8 quantization for smallest size
    nms=True,      # Include Non-Maximum Suppression to filter detections
    simplify=True,  # Simplify the model graph
)
```

```python
# Load the exported TFLite model
tflite_model = YOLO("yolo11n-pose_saved_model/yolo11n-pose_int8.tflite")

# Run inference with smaller input size
results = tflite_model("https://ultralytics.com/images/bus.jpg", imgsz=320)
```
